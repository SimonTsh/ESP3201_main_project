```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
%matplotlib notebook
```

```python
img = cv2.imread('frame500_mask.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_h, img_v = img_hsv[:,:,0], img_hsv[:,:,2]
```

```python
plt.figure(1); plt.imshow(img); plt.show()
```

```python
plt.figure(2); plt.imshow(img_h); plt.show()
```

```python
from enum import IntEnum

class SegClass(IntEnum):
    ROAD = 1
    BUILDING = 2
    FENCE = 3
    TRAFFIC_LIGHT = 4
    TRAFFIC_SIGN = 5
    PLANT = 6
    TERRAIN = 7
    SKY = 8
    CAR = 9

class SegID(IntEnum):
    ROAD = 7
    BUILDING = 11
    FENCE = 13
    TRAFFIC_LIGHT = 19
    TRAFFIC_SIGN = 20
    PLANT = 21
    TERRAIN = 22
    SKY = 23
    CAR = 26

img_out = np.zeros(img_h.shape, dtype=np.uint8)
img_out[(img_h > 27) & (img_h < 33)] = SegClass.CAR
img_out[(img_h > 57) & (img_h < 63)] = int(SegClass.PLANT)
img_out[(img_h > 85) & (img_h < 95)] = SegClass.TRAFFIC_SIGN
img_out[(img_h > 97) & (img_h < 103)] = SegClass.TRAFFIC_LIGHT
img_out[(img_h > 117) & (img_h < 123)] = SegClass.ROAD
img_out[(img_h > 147) & (img_h < 153)] = SegClass.TERRAIN

img_out[(img_v > 220) & (img_h < 3)] = SegClass.BUILDING
img_out[(img_v < 5) & (img_h < 3)] = SegClass.SKY

# img_out[(img_h > 10) & (img_h < 20)] = SegClass.FENCE
# fence ...

# buildings - blue
# obstacle - cyan
# sky - black
# green: 60 (plants)
# yellow: 90 (traffic signs)
# orange: 99-100 (traffic lights)
# red: 120 (roads)
# pink: 150 (ground)
plt.figure(3)
plt.imshow(img_out); plt.show()
```
